```r
wt = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv',
              col_types = cols()
              )


wt |> head()
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
p = ggplot(wt, aes(y = fct_infreq(province_territory))) +
    geom_bar()

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
wt_sorted = wt |>
    mutate(manufacturer = fct_infreq(manufacturer)) |>
    mutate(manufacturer = fct_lump(manufacturer, n = 5)) |>
    mutate(manufacturer = fct_relevel(manufacturer, 'Other', after = Inf)) |>
    mutate(manufacturer = fct_rev(manufacturer))

p = ggplot(wt_sorted, aes(y = manufacturer)) +
           geom_bar()
           
p
```
